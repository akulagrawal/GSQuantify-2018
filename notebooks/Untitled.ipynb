```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
for i in range(0,4):
    for j in range(0,4):
        plt.scatter(df[df.columns[i]],df[df.columns[j]])
        plt.xlabel(df.columns[i])
        plt.ylabel(df.columns[j])
        plt.show()
```

```python
domA = pd.DataFrame(df.loc[df['Domain'] == 'A'])
domB = pd.DataFrame(df.loc[df['Domain'] == 'B'])
domCi = pd.DataFrame(df.loc[df['Domain'] == 'C[i]'])
domCii = pd.DataFrame(df.loc[df['Domain'] == 'C[ii]'])
```

```python
print(domA.shape)
print(domB.shape)
print(domCi.shape)
print(domCii.shape)
```

```python
for i in range(0,3):
    for j in range(0,3):
        plt.scatter(domA[domA.columns[i]],domA[domA.columns[j]])
        plt.xlabel(domA.columns[i])
        plt.ylabel(domA.columns[j])
        plt.show()
```

```python
trainA = pd.DataFrame()
for k in range(0,120):
    trainA[k] = pd.Series(domA.shape[0]-119)
```

```python
for j in range(0,domA.shape[0]-119):
        for k in range(0,120):
            trainA[k][j] = domA['Usage'][domA.index[j+k]]
            trainA[k][j] = domA['InventoryCode'][domA.index[j+k]]
```
